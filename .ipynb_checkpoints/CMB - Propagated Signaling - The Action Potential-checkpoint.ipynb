# Propagated Signaling: The Action Potential

The action potential is caused by ion flux which depolarizs the cell membrane. Fast Na<sup>+</sup> influx depolarizes the cell; K<sup>+</sup> repolarizes it. Diversity of channels can alter the firing rate of neurons. Most of our knowledge comes from voltage clamping experiments and analysis of the molecular structure of channels. These insights can better tell us about genetic diseases stemming from ion channel changes and help us develop better, more effective drugs.

### The action potential is generated by the flow of ions through voltage-gated channels.

Cole and Curtis discovered in the 1940's that ion flux across the membrane changes at the action potential. Hodgkin and Katz found that the amplitude of the  AP is reduced when the sodium concentration is lowered (i.e. change in permeability drives the action potential). 

To study these permeability properties, Hodgkin and Huxley developed voltage clamping, which prevents action potentials from firing and disrupting the whole cell's permeability properties. It interrupts the intereaction between the membrane potential and ion channels opening by injecting opposite current into the axon.

Voltage clamp process:

1. Clamp $V_m$ at desired voltage (e.g. $V_R$).
2. Step voltage up to depolarize cell.
3. Step down to original voltage.
4. Manipulate process by using drugs such as TTX (block Na<sup>+</sup> channels) or TEA (K<sup>+</sup> channels).

#### Voltage-gated sodium and potassium conductances are calculated from their currents. 

#### The action potential can be reconstructed from the properties of sodium and potassium channels. 

Sodium and potassium channels have two similarities and two differences:

- Both open in response to depolarization.
- As the action potential rises in amplitude, both ions conductance increase.
- They differ in rate: sodium channels always open faster.
- They differ in response to prolonged depolarization: sodium channels close and only reactivate after raching a negative membrane voltage, while potassium channels do not close.

The "after-potential" is due to K<sup>+</sup> channels only closing *after* $V_m$ reaches the resting potential.

The **absolute refractory period** comes right after the action potential, when Na<sup>+</sup> channels are inactive and cannot be reopened. The **relative refractory period** comes when the membrane is still repolarizing but Na<sup>+</sup> channels are simply closed. Depolarization requires additional energy to overcome the rectifying processes.

The threshold is the point at which net flow of ions moves from outward to inwawrd (i.e. $I_{Na} + I_K + I_l = 0$). Small depolarizations can actually increase $I_K$ and $I_l$.

### Variations in the properties of voltage-gated ion cahnnels increase the signalling capabilities of neurons.

#### The nervous system expresses a rich variety of voltage-gated ion channels.4

A rich vvariety of ion channels adds to the neuron's ability to transmit complex signals. There are at least four major types of K<sup>+</sup> channels:

- Delayed rectifier: main one, discovered by Hodgkins and Huxley
- Calcium-activated
- A-type: acts like NaK<sup>+</sup> channel
- M-type: slowly activates at resting $V_m$, shut off at by ACh



#### Gating of voltage-sensitive ion channels can be influenced by various cytoplasmic factors.

Except for Ca<sup>2+</sup> ions, ion flux does not change the concentration of gradients. (e.g. Ca<sup>2+</sup> flow actually enhances the probability of K<sup>+</sup> gated channels.)

#### Excitability properties vary between regions of the neuron.

The trigger zone has the lowest threshold because of its high density of Na<sup>+</sup> ion channels. The axon is most like the squid giant axon: it mainly has fast Na<sup>+</sup> channels and leaky K<sup>+</sup> channels. Presynaptic terminals have many Ca<sup>2+</sup>.

#### Redistribution of charge within voltage-gated sodium channels controls channel gating.

#### The voltage-gated sodium channel selects forsodium on the basis of size, charge and energy of hydration of the ion.

Part of the inner wall has charged amino acid residues that can 'pop off' water molecules, and its effectiveness varies with different ions. Channel selectivity can be varied by subsituting different residues in the channel. 

### Genes encoding the potassium, sodium and calcium channels stem from a common ancestor.

#### Various smaller subunits contribute to the functional properties of voltage-gated sodium, calcium and potassium channels.

Na<sup>+</sup> channels have three subunits: a pore-forming $\alpha$, $\beta 1$ and $\beta 2$. There are four membrane-spanning domains in $\alpha$ subunit. These subunits can act as voltage sensors and can reconfigure in response to changes in membrane potential. They have modulatory functions of other subunits that may shift gating properties.

#### The diversity of voltage-gated channel types is due to several genetic mechanisms.

Different genes can encode for the same subunit; these various subunits can then combine to form slightly different channels. Different neurons cans also splice them differently.

#### Mutations in voltage-gated channels cause specific neurological diseases.

Point mutations can result in disease, such as Na<sup>+</sup> channels that failt to anactivate, causing muscle stiffness or paralysis. A variety of channels can be afflicted in these diseases.
